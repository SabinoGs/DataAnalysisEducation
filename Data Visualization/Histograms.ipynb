```python
import seaborn as sns
import numpy as np
import pandas as pd
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_set_1 = randn(100)
```

```python
plt.hist(data_set_1)
```

```python
"""
Mais informações sobre os argumentos de plt.hist:
https://matplotlib.org/devdocs/api/_as_gen/matplotlib.pyplot.hist.html

"""
data_set_2 = randn(80)
plt.hist(data_set_2,color='indianred')
```

```python
"""
Como juntar os 2 histogramas acima?

Precisamos somente chamar 2 vezes o plot. 
"""
plt.hist(data_set_1,normed=True,color='indianred',alpha=0.5,bins=20)
plt.hist(data_set_2,normed=True,color='blue',alpha=0.5,bins=20)
```

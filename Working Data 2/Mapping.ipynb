```python
"""
Metodo Map. 

Mapea valores com determinada condição
"""

regioes_brasil = {
    'AP':'Norte',
    'AM':'Norte',
    'PA':'Norte',
    'TO':'Norte',
    'RO':'Norte',
    'RR':'Norte',
    'AC':'Norte',
    'AL':'Nordeste',
    'BA':'Nordeste',
    'CE':'Nordeste',
    'MA':'Nordeste',
    'PB':'Nordeste',
    'PE':'Nordeste',
    'PI':'Nordeste',
    'RN':'Nordeste',
    'SE':'Nordeste'
    #Assim por diante
}


"""
Relacionar um dict que tem listas como values é mais dificil. 
"""
df['Regiões'] = df.UF.map(regioes_brasil)
df
```

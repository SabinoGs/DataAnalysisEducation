```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
arr = np.arange(6).reshape((2,3))
arr
```

```python
df = pd.DataFrame(arr,index=list("AB"),columns=['One','Two','Three'])
df
```

```python
"""
Realiza a soma de cada coluna.
"""
df.sum()
```

```python
"""
Retorna a média do DF
"""
df.mean()
```

```python
"""
Acha o menor valor de cada coluna

Existe o max() tbm
"""
df.min() 
```

```python
"""
Retorna em qual indice está o valor minimo de cada coluna. 
Existe o idxmax tambem
"""
df.idxmin()
```

```python
"""
Soma cumulativa.
"""
df.cumsum()
```

```python
"""
Retorna um sumário de estatisticas do DF.
"""
df.describe()
```

```python
"""
Pegar dados do mercado Financeiro.


"""

import pandas_datareader as pdr
import datetime
```

```python
stocks = pdr.get_data_yahoo(['GOOG','AMZN'],start=datetime.datetime(2010,1,1),
                            end=datetime.datetime(2017,1,1))
```

```python
prices = stocks['Adj Close']
prices
```

```python
volume = stocks['Volume']
volume
```

```python
rets = prices.pct_change()
rets
```

```python
rets.corr()
```

```python
prices.plot()
```
